```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/Users/aharo/desk/python/machine_learning/zcode/guidance/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df,y='reading score',x='parental level of education', hue='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.boxplot(data=df,x='reading score',y='parental level of education', hue='test preparation course', palette='Set2')
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.violinplot(data=df,x='reading score',y='parental level of education', palette='Set2')
#sns.violinplot(data=df,x='reading score',y='parental level of education', hue='test preparation course', palette='Set2') # in violins we usually dont do hue's 
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.swarmplot(data=df,x='math score',y='gender', hue='test preparation course',dodge=True) 

plt.legend(bbox_to_anchor=(1.355,0.5))
```
