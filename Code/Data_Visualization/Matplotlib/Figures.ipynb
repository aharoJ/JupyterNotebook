```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# DATA
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x= np.arange(0,10)
```

```python
y= 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
#Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_title("aharo24")
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel('B')



#Zoomed in  Axes
axes2 = fig.add_axes([0.2,0.5,0.5,0.1])
axes2.plot(x,y)


axes2.set_title("ZOOM")
axes2.set_xlim(1,6)
axes2.set_ylim(0,30)
axes2.set_xlabel("A")
axes2.set_ylabel('B')

```
